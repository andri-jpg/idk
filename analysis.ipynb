```python
import pandas as pd

# Path ke file CSV
csv_file_path = "virus.csv"

# Baca file CSV menjadi DataFrame Pandas
df = pd.read_csv(csv_file_path, sep='|')

# Konversi DataFrame menjadi list Python
csv_data = df.values.tolist()

# Ubah isi file CSV menjadi hardcoded data dalam variabel Python
hardcoded_data = "[" + ",\n".join(str(row) for row in csv_data) + "]"

# Tampilkan hardcoded data
print(hardcoded_data)
```
